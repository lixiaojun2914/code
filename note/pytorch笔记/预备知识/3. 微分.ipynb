```python
import torch
import numpy as np
from d2l import torch as d2l
```

```python
def f(x):
    return 3 * x ** 2 - 4 * x
```

```python
def numerical_lim(f, x, h):
    return (f(x + h) - f(x)) / h


h = 0.1
for i in range(5):
    print(f'h={h:.5f}, numerical limit={numerical_lim(f, 1, h):.5f}')
```

```python
x = np.arange(0, 3, 0.1)
d2l.plot(x, [f(x), 2 * x - 3], 'x', 'f(x)', legend=['f(x)', 'Tangent line (x=1)'])
```

```python
x = torch.arange(4.0, requires_grad=True)
x.grad
```

```python
y = 2 * torch.dot(x, x)
y.backward()
x.grad
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```
